```python
import re

data = open('input4.txt').read().split('\n\n')


def isValid(item):
    colors = ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    string = item.split()
    if len(string) >= 7:
        for i in string:
            i = i.split(":")
            a = i[0]
            b = i[1]
            if len(string) == 7:
                if a == 'cid':
                    return False
            if a == 'byr':
                if len(b) != 4 or not (1920 <= int(b) <= 2002):
                    print(i)
                    return False
            elif a == 'iyr':
                if len(b) != 4 or not (2010 <= int(b) <= 2020): 
                    print(i)
                    return False
            elif a == 'eyr':
                if len(b) != 4 or not (2020 <= int(b) <= 2030): 
                    print(i)
                    return False
            elif a == 'hgt':
                match = re.match("([0-9]+)([a-z]+)", b)
                if match:
                    number, letter = match.groups()
                    if letter == 'cm' and 150 <= int(number) <= 193: continue
                    elif letter == 'in' and 59 <= int(number) <= 76: continue
                    else: 
                        print(i)
                        return False
                else:
                    print(i)
                    return False
            elif a == 'hcl':
                if len(b) != 7 or b[0] != '#' or not(re.match("^[a-f0-9]*$", b[1:])):
                    print(i)
                    return False
            elif a == 'ecl':
                if b not in colors: 
                    print(i)
                    return False
            elif a == 'pid':
                if len(b) != 9: 
                    print(i)
                    return False
        return True
    else: return False

    
sum = 0
for i in data:
    print('new')
    if isValid(i): sum += 1

sum
```
