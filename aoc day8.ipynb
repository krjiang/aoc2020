```python
data = open('input8.txt').read().strip().split('\n')

def accumulate(index, accumulator):
    if index >= len(data):
        print('done')
        print(accumulator)
        return accumulator
    if index in indices:
        return
    indices.append(index)
    line = data[index]
    command = line.split()[0]
    value = int(line.split()[1])
    if command == 'nop':
        return accumulate(index + 1, accumulator)
    elif command == 'acc':
        accumulator += value
        return accumulate(index + 1, accumulator)
    elif command == 'jmp':
        return accumulate(index + value, accumulator)
```

```python
for index in range(len(data)):
    indices = []
    line = data[index]
    command = line.split()[0]
    value = line.split()[1]
    if command == 'nop':
        data[index] = 'jmp ' + value
    elif command == 'jmp':
        data[index] = 'nop ' + value
    accumulate(0,0)
    data[index] = command + ' ' + value
```
